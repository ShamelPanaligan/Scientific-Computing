# De Moivre's Thoerem Explored
In this notebook, we explore De Moivre by verifying through the comparison of a different method. We experiment with visual results to see the effect of powers of complex numbers and lastly, we look over the roots of unity.

```python
import cmath
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
```

## Verfication Of De Moivre
Comparing the De Moivre method with Direct Multiplication Method

```python
# Converts complex number to polar form
# Let z be a complex number
z = 1 + 2j 
def complex_polar(z):
    import cmath
    r = abs(z)
    theta = cmath.phase(z)
    return r, theta


r, theta = complex_polar(z)


print("*****************************************")
print("Direct Multiplication of Complex Numbers")
# Using z1 = 1 + 2j and z2 = 3 + 4j
z1 = complex(1, 2)
z2 = complex(3, 4)

Real_part = z1.real * z2.real - z1.imag * z2.imag
Imag_part = z1.real * z2.imag + z1.imag * z2.real
z_product = complex(Real_part, Imag_part)
print(f"z3 = z1 * z2 = {z_product} (Direct Multiplication)")
print()
print("*****************************************")
print("Multiplication Using De Moivre's Theorem")
r1, theta1 = complex_polar(z1)
r2, theta2 = complex_polar(z2)
r3 = r1 * r2
theta3 = theta1 + theta2
z_product_de_moivre = r3 * (cmath.cos(theta3) + 1j * cmath.sin(theta3))
print(f"z3 = z1 * z2 = {z_product_de_moivre} (Using De Moivre's Theorem)")
print()
assert np.isclose(z_product.real, z_product_de_moivre.real) and np.isclose(z_product.imag, z_product_de_moivre.imag), "Results do not match!"
print("Both methods yield the same result.")
```

## Powers of Complex Numbers

```python
# Showing the powers of n and how it affects the scaling and angle of the plot
print("Powers of z using De Moivre's Theorem:")
print("As a reminder, De Moivre's Theorem states that for a complex number in polar form:")
print("z = r(cos θ + j sin θ)")
print("Then z^n = r^n(cos nθ + j sin nθ)")

plt.figure(figsize=(8, 8))
plt.xlabel('Real axis')
plt.ylabel('Imaginary axis')
plt.title('Argand Diagram')
plt.axis('equal')
plt.xlim(-50, 50)
plt.ylim(-50, 50)

for n in range(1, 6):
    r_n = r ** n
    theta_n = n * theta
    cos_theta_n = np.cos(theta_n)
    sin_theta_n = np.sin(theta_n)
    complex_n = r_n * (cmath.cos(theta_n) + 1j * cmath.sin(theta_n))
    plt.plot([0, r_n * cos_theta_n], [0, r_n * sin_theta_n], marker='o', label=f'z^{n}')
    plt.text(r_n * cos_theta_n, r_n * sin_theta_n, f'z^{n}', fontsize=12, ha='right')

plt.show()
```

### As you can see, it forms a logarithmic spiral, as the theta increases linearly and r increases exponentially. Then, a similar yet reverse effect can be seen when raising to negative powers.

```python
# Negative Powers of z using De Moivre's Theorem
print("Negative Powers of z using De Moivre's Theorem:")
plt.figure(figsize=(8, 8))
plt.xlabel('Real axis')
plt.ylabel('Imaginary axis')
plt.title('Argand Diagram')
plt.axis('equal')
plt.xlim(-1, 1)
plt.ylim(-1, 2.5)

for n in range(1, -4, -1):
    r_n = r ** n
    theta_n = n * theta
    cos_theta_n = np.cos(theta_n)
    sin_theta_n = np.sin(theta_n)
    complex_n = r_n * (cmath.cos(theta_n) + 1j * cmath.sin(theta_n))
    plt.plot([0, r_n * cos_theta_n], [0, r_n * sin_theta_n], marker='o', label=f'z^{n}')
    plt.text(r_n * cos_theta_n, r_n * sin_theta_n, f'z^{n}', fontsize=7, ha='right')

plt.show()
```

## Roots of Unity
Finding n-th roots using De Moivre
<br> We want z so that z^n = 1
<br> By De Moivre: rⁿ(cos(nθ) + i sin(nθ)) = 1(cos0 + i sin0)
<br> thus r^n = 1 implies r = 1 (meaning we expect all roots to lie on a unit circle)
<br> Since n * theta = theta + 2 * pi * k, for int k, rearrange to get theta = 2 * pi * k / n



```python
# For z^n =1
print("Finding the roots of unity for z^n = 1:")
# let n =5

n = 5
roots = []

# Create the axis
plt.figure(figsize=(8, 8))
plt.xlabel('Real axis')
plt.ylabel('Imaginary axis')
plt.title(f'Roots of Unity for z^{n} = 1')
plt.axis('equal')
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
circle = plt.Circle((0, 0), 1, color='lightgray', fill=False, linestyle='dashed')
plt.gca().add_artist(circle)

#Calculate and plot the roots
for k in range(n):
    theta_k = 2 * np.pi * k / n
    root_k = cmath.cos(theta_k) + 1j * cmath.sin(theta_k)
    roots.append(root_k)
    plt.plot([0, root_k.real], [0, root_k.imag], marker='o', label=f'Root {k+1}')
    plt.text(root_k.real, root_k.imag, f'R{k+1}', fontsize=12, ha='right')  
```
